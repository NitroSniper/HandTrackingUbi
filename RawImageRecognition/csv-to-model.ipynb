```python
import pickle

import pandas as pd

df = pd.read_csv('hand_landmarks_with_class.csv')
```

```python
label_col = "Class"
SEED = 42
```

```python
from sklearn.preprocessing import LabelEncoder
X, y = df.drop(label_col, axis=1), df[label_col].copy()
label = LabelEncoder()
X_processed = X
y_processed = label.fit_transform(y)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_processed, y_processed,
                                                    test_size=0.2, random_state=SEED,
                                                    stratify=y_processed)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_train[:2], y_train[:2], label.classes_
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_preds = neigh.predict(X_test)
print(accuracy_score(y_test, y_preds))
print(confusion_matrix(y_test, y_preds))
```

```python
import pickle
pickle.dump((neigh, label), open('neigh-model.pkl', 'wb'))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Input
from tensorflow.keras.models import Sequential


class RelativePositionLayer(layers.Layer):
    def call(self, inputs):
        # Reshape from (63,) to (21, 3)
        coords = tf.reshape(inputs, (-1, 21, 3))
        diffs = tf.expand_dims(coords, 2) - tf.expand_dims(coords, 1)
        distances = tf.norm(diffs, axis=-1)  # Euclidean distances
        return distances  # shape: (21, 21)
```

```python
num_classes = len(label.classes_)
model = Sequential([
    Input((63,)),
    RelativePositionLayer(),
    layers.Flatten(),                      # flatten to (441,)
    layers.Dense(128, activation='relu'),  # fully connected layer
    layers.Dense(64, activation='relu'),   # another layer
    layers.Dense(num_classes, activation='softmax')  # output layer
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
X_train.shape
```

```python
epochs = 10

history = model.fit(
    x=X_train,
    y=y_train,
    validation_data=0.2,
    epochs=epochs
)
```
