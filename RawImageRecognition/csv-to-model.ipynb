```python
import pickle
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
df = pd.read_csv('hand_landmarks_with_class.csv')
df.head()
```

```python
df.info()
```

```python
label_col = "Class"
SEED = 42
```

```python
X, y = df.drop(label_col, axis=1), df[label_col].copy()
label = LabelEncoder()
X_processed = X
y_processed = label.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y_processed,
                                                    test_size=0.2, random_state=SEED,
                                                    stratify=y_processed)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
X_train[:2], y_train[:2], label.classes_
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
y_preds = neigh.predict(X_test)
print(accuracy_score(y_test, y_preds))
print(confusion_matrix(y_test, y_preds))
```

```python
pickle.dump((neigh, label), open('neigh-model.pkl', 'wb'))
```
